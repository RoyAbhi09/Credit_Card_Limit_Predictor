```python
import numpy as np
import pandas as pd
```

```python
d=pd.read_csv('/kaggle/input/credit-card/credit.csv')
d.head(10)
```

```python
d1 = pd.get_dummies(d['Gender'], drop_first=True).astype(int) 
d = d.drop('Gender', axis=1)
d = pd.concat([d, d1], axis=1)

```

```python
#For Student Section
d['Student'] = d['Student'].map({'Yes': 1, 'No': 0})
#For Married Section
d['Married'] = d['Married'].map({'Yes': 1, 'No': 0})
# For Ethnicity
d['Ethnicity'] = d['Ethnicity'].map({'Caucasian': 1, 'Asian': 2,'African American':3})
```

```python
d.head(10)
```

```python
X=d.iloc[:,[0,2,3,4,5,6,7,8,9,10]]
Y=d.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
m=RandomForestRegressor(n_estimators=100)
m.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=m.predict(X_test)
acc=r2_score(y_pred,y_test)
acc*100.0
```
